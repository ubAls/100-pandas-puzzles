```python
from glob import glob

train_label_list = glob('/content/yolov5/train/labels/*.txt')
for file in train_label_list:
  print(file)
  f = open(file, "rt")
  person = []
  for line in f:
    if line.startswith("48"):
      person.append(line)
  f.close()

  f2 = open(file, 'wt')
  for p in person:
    f2.write(p)
  f2.close()
  f3 = open(file, "r")
  for line in f3:
    print(line)
  f3.close()
  break
```

```python
from glob import glob
import os
train_label_list = glob('/content/yolov5/train/labels/*.txt')
for path in train_label_list:
  print(path)
  f = open(path, "rt")
  person = []
  for line in f:
    if line.startswith("48"):
      person.append(line)
  f.close()
  if len(person) == 0 :
    os.remove(path)
    os.remove(path.replace(".txt", ".jpg").replace("/labels/","/images/"))
  else :
    f2 = open(path, 'wt')
    for p in person:
      f2.write(p.replace("48","0",1))
    f2.close()
```

```python
import os
from PIL import Image

# 데이터셋 경로
dataset_path = '/content/my_drive/MyDrive/datasets/data/datasets'

# 결과를 저장할 디렉터리 생성
output_path = '/content/my_drive/MyDrive/datasets/data/label'
os.makedirs(output_path, exist_ok=True)

# 통일된 라벨
uniform_label = "1 0.7966145833333333, 0.700925925925926, 0.3984375, 0.9222222222222223"

# 이미지 파일 목록
image_files = [f for f in os.listdir(dataset_path) if f.endswith(('.jpg', '.jpeg', '.png'))]

# tqdm을 사용하여 진행 상황 시각화
for image_file in image_files:
    # 이미지 파일 경로
    image_path = os.path.join(dataset_path, image_file)

    # 라벨 파일 경로
    label_path = os.path.join(output_path, f"{os.path.splitext(image_file)[0]}.txt")

    # 통일된 라벨 쓰기
    with open(label_path, 'w') as label_file:
        label_file.write(uniform_label)

print("라벨링된 데이터 생성 완료.")
```

```python
def convert_to_yolo_format(top, left, height, width, image_width, image_height):
    # 중심 좌표 계산
    x_center = left + width / 2
    y_center = top + height / 2

    # YOLO 형식으로 정규화
    x_center /= image_width
    y_center /= image_height
    width /= image_width
    height /= image_height

    return x_center, y_center, width, height

# 주어진 좌표 값
top, left, height, width = 259, 1147, 996, 765

# 전체 이미지 크기 (예: 이미지의 가로 크기 1920, 세로 크기 1080)
image_width, image_height = 1920, 1080

# YOLO 형식으로 변환
yolo_format = convert_to_yolo_format(top, left, height, width, image_width, image_height)

print("YOLO 형식으로 변환된 좌표:", yolo_format)
```
